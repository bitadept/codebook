```python
import pandas as pd
```

```python
df = pd.read_csv('weather_data.csv', parse_dates = ['day'])
df.set_index('day', inplace = True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
new_df = df.fillna(method = 'ffill')
new_df
```

```python
new_df = df.fillna(method = 'bfill', axis = 'rows', limit = 2)
new_df
```

```python
new_df = df.interpolate(method = 'time')
new_df
```

```python
new_df = df.dropna(how = 'all')
new_df
```

```python
new_df = df.dropna(thresh = 2)
new_df
```

```python
dt = pd.date_range('01-01-2017', '01-11-2017')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
