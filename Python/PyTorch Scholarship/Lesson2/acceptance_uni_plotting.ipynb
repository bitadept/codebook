```python
%matplotlib widget
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.style.use(['dark_background'])
```

```python
NUM_STUDENTS = 180

TEST_WEIGHT = 2
GRADES_WEIGHT = 1
BIAS = 18

TEST = 'test'
GRADES = 'grades'
ACCEPTED = 'accepted'

fig, axs = plt.subplots()

student_data = []
for i in range(NUM_STUDENTS):
    test = round(np.random.uniform(10), 2)
    grades = round(np.random.uniform(10), 2)
    student_data.append({TEST: test, GRADES: grades, ACCEPTED: (TEST_WEIGHT * test + GRADES_WEIGHT * grades) >= BIAS})

# print(student_data)

for student in student_data:
    color = 'bo' if student[ACCEPTED] == True else 'ro'
    axs.plot(student[TEST], student[GRADES], color)
    
x = np.linspace(0, 12, 2)
axs.plot(x, (TEST_WEIGHT * x - BIAS) / -GRADES_WEIGHT)
```
